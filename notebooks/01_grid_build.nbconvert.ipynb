# Notebook 01: Build Geospatial Grid

Constructs a 500 m × 500 m grid over the Delhi bounding box and saves it as a GeoJSON file.

```python
import sys
sys.path.insert(0, '..')

from src.grid_builder import build_grid, save_grid
```

## Build the grid

Delhi bounding box: `(76.8, 28.4, 77.4, 28.9)` with 500 m cells.

```python
bbox = (76.8, 28.4, 77.4, 28.9)  # Delhi bounding box
gdf = build_grid(bbox, cell_size_meters=500)
print(f"Total cells generated: {len(gdf)}")
gdf.head()
```

## Save grid to disk

```python
import os
os.makedirs('../data/processed', exist_ok=True)
save_grid(gdf, '../data/processed/grid.geojson')
print("Grid saved successfully.")
```

## Quick visual check

```python
import matplotlib
matplotlib.use('Agg')          # non-interactive backend (notebook safe)
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(ax=ax, edgecolor='steelblue', facecolor='lightcyan', linewidth=0.3)
ax.set_title('Delhi 500m Grid', fontsize=14)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.tight_layout()
plt.show()
```
